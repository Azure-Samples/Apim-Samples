### 🛠️ Configure Infrastructure Parameters & Create the Infrastructure

Set your desired parameters for the AG-APIM-VNET infrastructure deployment.

❗️ **Modify entries under _User-defined parameters_**.

🔐 **Important**: After deployment, you'll need to install a self-signed certificate to avoid SSL warnings when testing HTTPS endpoints. This can be done using the Python cell below.

```python
import utils
from apimtypes import *

# ------------------------------
#    USER CONFIGURATION
# ------------------------------

rg_location = 'eastus2'             # Azure region for deployment
index       = 1                     # Infrastructure index (use different numbers for multiple environments)
apim_sku    = APIM_SKU.DEVELOPER    # Options: 'DEVELOPER', 'STANDARDV2', 'PREMIUMV2'



# ------------------------------
#    SYSTEM CONFIGURATION
# ------------------------------

inb_helper = utils.InfrastructureNotebookHelper(rg_location, INFRASTRUCTURE.AG_APIM_VNET, index, apim_sku) 
inb_helper.create_infrastructure()

utils.print_ok('All done!')
```

### 🔐 Install Self-Signed Certificate (REQUIRED for HTTPS Testing)

⚠️ **This step is required to avoid SSL certificate warnings when testing Application Gateway HTTPS endpoints.**

The deployment creates an HTTPS-only Application Gateway with a self-signed certificate. To browse to the endpoints without SSL warnings, you must install this certificate in your local trust store.

#### Installation Method: Python (Cross-Platform)
Use the Python cell below to automatically install the certificate.

#### How it works:
- A self-signed certificate with CN `apim-samples-{resourceGroupName}` is created during deployment
- The certificate is stored securely in Azure Key Vault
- The installation process downloads and installs it to your local trusted root certificate store
- The certificate is tagged with the "apim-samples-" prefix for easy identification and cleanup

```python
# Install Certificate using Python (Cross-Platform)
# This installs the self-signed certificate to your local trust store

utils.print_info("🔐 Installing self-signed certificate...")
utils.print_info(f"Certificate name: apim-samples-{inb_helper.resource_group_name}")

success = utils.install_certificate_for_infrastructure(
    INFRASTRUCTURE.AG_APIM_VNET, 
    index,
    inb_helper.resource_group_name
)

if success:
    utils.print_ok("✅ Certificate installation completed!")
    utils.print_info("You can now browse to HTTPS endpoints without SSL warnings.")
else:
    utils.print_error("❌ Certificate installation failed.")
    utils.print_info("💡 Note: Certificate automatically installed for HTTPS endpoint testing")
```

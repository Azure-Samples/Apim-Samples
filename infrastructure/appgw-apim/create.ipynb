### 🛠️ Configure Infrastructure Parameters & Create the Infrastructure

Set your desired parameters for the APPGW-APIM infrastructure deployment.

❗️ **Modify entries under _User-defined parameters_**.

**Note:** This infrastructure includes Application Gateway with API Management using VNet injection.

**NSG model:** `nsg-default` is the generic fallback. When `use_strict_nsg` is `False`, service subnets use permissive service-aware NSGs (`nsg-appgw`, `nsg-apim`, and optionally `nsg-aca`). When `use_strict_nsg` is `True`, those subnets switch to the restrictive `-strict` variants.

```python
from apimtypes import APIM_SKU, INFRASTRUCTURE, Region
from utils import InfrastructureNotebookHelper
from console import print_ok

# ------------------------------
#    USER CONFIGURATION
# ------------------------------

rg_location = Region.EAST_US_2      # Azure region for deployment
index       = 1                     # Infrastructure index (use different numbers for multiple environments)
apim_sku    = APIM_SKU.DEVELOPER    # Options: 'DEVELOPER', 'PREMIUM', 'PREMIUMV2' (VNet injection supports Developer, Premium, and PremiumV2)



# ------------------------------
#    SYSTEM CONFIGURATION
# ------------------------------

use_strict_nsg = False              # Optional: deploy strict NSGs for supported subnets. Disabled by default to keep the sample focused.

inb_helper = InfrastructureNotebookHelper(
    rg_location,
    INFRASTRUCTURE.APPGW_APIM,
    index,
    apim_sku,
    use_strict_nsg,
)
inb_helper.create_infrastructure()

print_ok('All done!')
```
